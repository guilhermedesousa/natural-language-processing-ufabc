#### **INTRODUÇÃO**

**Descrição da API**

A brapi é uma Interface de Programação de Aplicativos (API) que fornece informações sobre ações, moedas e criptomoedas.

Ela fornece acesso a uma ampla gama de dados relacionados a ações, preço atual, preço mínimo e máximo do dia ou ano, variações do preços, o volume de negociações, entre outros. Além disso, para moedas e criptomoedas, é possível fazer conversões entre os diferentes tipos de moedas.

A API é útil para construção de aplicações web e nativas que necessitem exibir esses dados de forma centralizada para os seus usuários. Ela oferece uma resposta rápida à requisições, tem uma infraestrutura construída em nuvem e possui alta disponibilidade.

Atualmente a API possui alguns planos, sendo que o plano "GRATUITO" é ideal para testar a API ou para projetos pessoais. O plano "STARTUP" possui alguns recursos a mais, como dados mais recentes, e é ideal para apps e sites com tráfego moderado. Por fim, o plano "PRO" permite um número maior de requisições e é ideal para machine learning, apps e sites com tráfego intenso.

#### **AUTENTICAÇÃO**

Esta API oferece autenticação para garantir que apenas usuários autorizados possam acessar os recursos.

A autenticação deve ser feita usando um token como parâmetro de consulta (query param) na URL da seguinte maneira:

```
https://brapi.dev/api/{endpoint}?token={seu_token}
```

Basta substituir o `{seu_token}` pelo token de autenticação fornecido a você.

**Como encontrar seu Token**

Para encontrar o seu token de autenticação, siga os seguinte passos:

1.   Crie uma conta na brapi no seguinte [link](https://brapi.dev/dashboard)
2.   Acesse a [Dashboard da brapi](https://brapi.dev/dashboard) após o registro
3.   Em "Seus tokens", clique em "Criar novo token"

Agora, basta copiar o token criado e utilizá-lo nas requisições. Para isso armazene o token em uma variável:

````
token = 'seu_token'
````

#### **UTILIZAÇÃO BÁSICA**

**Como fazer solicitações**

O acesso aos dados financeiros é feito através de requisições HTTP simples. Você pode realizar solicitações do tipo GET para acessar uma variedade de informações financeiras. A estrutura básica de uma solicitação é a seguinte:

```
https://brapi.dev/api/{endpoint}
```

Substitua `{endpoint}` pelo endpoint específico que você queira acessar. Por exemplo, se você quiser saber informações detalhadas sobre cotações de ações, fundos, índices e BDRs, basta utilizar o seguinte endpoint:

```
https://brapi.dev/api/quote/list?token={seu_token}
```

Substitua `{token}` pelo seu token obtido no passo anterior.

**Como interpretar as respostas**

As respostas da API são retornadas em formato JSON, tornando-as fáceis de serem processadas por aplicativos e serviços.

Para isso, você pode utilizar a biblioteca `json` do Pyhton, que fornece uma interface simples para ler os dados da resposta e torná-los num formato manipulável pelo Python.

Para demonstrar o uso da API, crie uma nova célula de código e execute o seguinte código para obter as cotações de todas as ações, fundos, índices e BDRs.

````
endpoint = '/quote/list'
url = f"{api_base_url}{endpoint}?token={token}"

response = requests.get(url)

if response.status_code == 200:
  data = response.json()
  print(data)
else:
    print('Erro na requisição:', response.status_code)
    print(f"Error message: {response.text}")
````

#### **EXEMPLOS PRÁTICOS**

**Obter informação sobre uma ação específica**

A seguinte célula de código cria uma função que utiliza o endpoint `{/quote/tickers}` para buscar a cotação algum ativo. Ela obtém dados atualizados e precisos sobre este ativo.

Nesse exemplo, são utilizados alguns parâmetros na requisição, incluindo:

*   `range`: intervalo de tempo para os dados solicitado. Por exemplo: `5d` (5 dias)
*   `interval`: intervalo de tempo entre os dados. Por exemplo: `1d` (1 dia)
*   `fundamental`: um valor booleano que determina se você deseja dados fundamentais. Por exemplo: `true`
*   `dividends`: um valor booleano que termina se você deseja buscar informações sobre dividendos: Por exemplo: `false`

````
def requisicaoweb(tickers):
  url_q = api_base_url + '/quote/' + tickers

  params = {
    'range': '5d',
    'interval': '1d',
    'fundamental': 'true',
    'dividends': 'false',
    'token': token,
  }

  try:
    requisicao = requests.get(url_q, params=params)
    requisicao.raise_for_status()  # Verificar se há erros na requisição

    data = requisicao.json()
    print(data)

    simbolo = data['results'][0]['symbol']
    preco = data['results'][0]['regularMarketPrice']
    moeda = data['results'][0]['currency']
    valor_mercado = data['results'][0]['marketCap']
    alta_dia = data['results'][0]['regularMarketDayHigh']
    baixa_dia = data['results'][0]['regularMarketDayLow']
    abertura = data['results'][0]['regularMarketOpen']
    baixa_52_semanas = data['results'][0]['fiftyTwoWeekLow']
    alta_52_semanas = data['results'][0]['fiftyTwoWeekHigh']

    print(f"Símbolo: {simbolo}.")
    print(f"Moeda: {moeda}.")
    print(f"Preco: {preco}")
    print(f"Valor de Mercado: {valor_mercado}.")
    print(f"Alta do Dia no Mercado Regular: {alta_dia}.")
    print(f"Baixa do Dia no Mercado Regular: {baixa_dia}.")
    print(f"Abertura do Mercado Regular: {abertura}.")
    print(f"Baixa do Ano (52 semanas): {baixa_52_semanas}.")
    print(f"Alta do Ano (52 semanas): {alta_52_semanas}.")

    if 'updatedAt' in data['results'][0]:
      atualizado_em = data['results'][0]['updatedAt']
      print(f"Atualizado em: {atualizado_em}")

  except requests.RequestException as e:
      print(f'Erro na requisição: {e}')
      print(f"Error message: {requisicao.text if 'requisicao' in locals() else ''}")
````

**Obter informação sobre a taxa SELIC**

A seguinte célula de código cria uma função que utiliza o endpoint `{/v2/prime-rate}` para buscar informações sobre a taxa SELIC desde 01/12/2021 até a data atual.

Nesse exemplo, são utilizados alguns parâmetros na requisição, incluindo:

*   `country`: país que será retornado a taxa SELIC.
*   `start`: data de início para obtenção dos dados.
*   `end`: data final para obtenção dos dados utilizando a biblioteca datetime para retornar o dia atual.
*   `sortBy`: parâmetro para definir por qual tipo de dado o retorno será ordenado, neste caso por data.
*   `sortOrder`: parâmetro para definir a ordem que será ordenado, neste caso em ordem crescente por data.

````
def consultarSELIC():
  url_s = api_base_url + "/v2/prime-rate"

  data_atual = datetime.now()
  data_atual_formatada = str(data_atual.strftime('%d/%m/%Y'))

  params = {
    'country': 'brazil',
    'start': '01/12/2021',
    'end': data_atual_formatada,
    'sortBy': 'date',
    'sortOrder': 'asc',
    'token': token,
  }

  requisicao = requests.get(url_s, params=params)
  if requisicao.status_code == 200:
    dados = requisicao.json()
    valores = [float(entry['value']) for entry in dados['prime-rate']]
    dias = [entry['date'] for entry in dados['prime-rate']]

    quantidade_de_datas = 50
    indices = np.round(np.linspace(0, len(dias) - 1, quantidade_de_datas)).astype(int)

    # Selecionando os dias e valores correspondentes aos índices escolhidos
    dias_selecionados = [dias[i] for i in indices]
    valores_selecionados = [valores[i] for i in indices]

    plt.figure(figsize=(10, 3))
    plt.plot(dias_selecionados, valores_selecionados, marker='o', linestyle='-')
    plt.title('Variação da taxa SELIC ao longo do tempo')
    plt.xlabel('Data')
    plt.ylabel('Valor')
    plt.xticks(rotation=45)
    plt.grid(True)

    plt.tight_layout()
    plt.show()
    time.sleep(5)
    plt.close()
  else:
      print(f"Request failed with status code {requisicao.status_code}")
      print(f"Error message: {requisicao.text}")
````

````
consultarSELIC()
````

Imagine que você gostaria de trabalhar com importação de produtos estrangeiros. Logicamente, faz-se necessário saber a cotação atual da moeda para que sejam realizados os trâmites cambiais. Vamos utilizar o sufixo do URL "/v2/currency" (sem aspas) para que a API retorne o valor da moeda estrangeira com base na última atualização do dia. Para isso, precisamos passar como parâmetro tal moeda. No exemplo abaixo, utilizamos USD, que se refere à dólares estadunidenses.

````
def consultarMoedas(relacao):
  url_m = api_base_url + "/v2/currency"
  params = {
    'currency': relacao,
    'token': token,
  }

  response = requests.get(url_m, params=params)

  if response.status_code == 200:
    data = response.json()
    dados_relacao = data['currency']

    moeda_pais_origem = dados_relacao[0]['fromCurrency']
    moeda_pais_destino = dados_relacao[0]['toCurrency']
    nome_moedas = dados_relacao[0]['name'].split('/')
    valor_de_mercado = dados_relacao[0]['bidPrice']
    data = dados_relacao[0]['updatedAtDate']

    print(f"Moeda do pais origem: {nome_moedas[0]} ({moeda_pais_origem}).")
    print(f"Moeda do pais destino: {nome_moedas[1]} ({moeda_pais_destino}).")
    print(f"1 {nome_moedas[0]} equivale a {valor_de_mercado} {nome_moedas[1]}.")
    print(f"data: {data}")
  else:
      print(f"Request failed with status code {response.status_code}")
      print(f"Error message: {response.text}")
````
